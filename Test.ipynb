# 0. Получение данных

````python
import requests
import json
import pandas as pd
````

````python
result = requests.get("https://swapi.dev/api/planets/?search=Tatooine")
result.content
y = json.loads(result.content)
chars_links = y['results'][0]['residents']
chars_links
````

````python
def get_info(l):
    res = requests.get(l)
    return json.loads(res.content)
````

````python
def fill_info(d, nship, hworld = "Tatooine"):
    info = {
            'name': str(d['name']),
            'homeworld.name': str(hworld),
            'height': int(d['height']),
            'gender': str(d['gender']),
            'starships.name': str(nship)
        }
    return info
````

````python
characters_list = []
all_data = []
for link in chars_links:
    char_dict = get_info(link)
    
    all_data.append(char_dict)
    
    if len(char_dict['starships']) == 0:
        char_info = fill_info(char_dict, None)
        characters_list.append(char_info)
    else:
        for ship_link in char_dict['starships']:
            shipname = get_info(ship_link)['name']
            char_info = fill_info(char_dict, shipname)
            characters_list.append(char_info)
        
characters_df = pd.DataFrame(characters_list)
````

### Создание БД

````python
import mysql.connector
from mysql.connector import errorcode
````

````python
def create_database(cursor):
    try:
        cursor.execute(
            "CREATE DATABASE {} DEFAULT CHARACTER SET 'utf8'".format(DB_NAME))
    except mysql.connector.Error as err:
        print("Failed creating database: {}".format(err))
        exit(1)
````

````python
def create_tables(tables): 
    cnx = mysql.connector.connect(user='USER', database='characters_db')
    cursor = cnx.cursor()
    for table_name in tables:
        table_description = tables[table_name]
        try:
            print("Creating table {}: ".format(table_name), end='')
            cursor.execute(table_description)
        except mysql.connector.Error as err:
            if err.errno == errorcode.ER_TABLE_EXISTS_ERROR:
                print("already exists.")
            else:
                print(err.msg)
        else:
            print("OK")

    cursor.close()
    cnx.close()
````

````python
DB_NAME = 'characters_db'

TABLES = {}
TABLES['characters'] = (
    "CREATE TABLE `characters` ("
    "  `no` int(11) NOT NULL AUTO_INCREMENT,"
    "  `name` varchar(30) NOT NULL,"
    "  `homeworld_name` varchar(14) NOT NULL,"
    "  `height` int(5) NOT NULL,"
    "  `gender` enum('male','female') NULL,"
     "  `starships_name` varchar(30) NULL,"
    "  PRIMARY KEY (`no`)"
    ") ENGINE=InnoDB")

cnx = mysql.connector.connect(user='root')
cursor = cnx.cursor()
try:
    cursor.execute("USE {}".format(DB_NAME))
except mysql.connector.Error as err:
    print("Database {} does not exists.".format(DB_NAME))
    if err.errno == errorcode.ER_BAD_DB_ERROR:
        create_database(cursor)
        print("Database {} created successfully.".format(DB_NAME))
        cnx.database = DB_NAME
    else:
        print(err)
        exit(1)
````

````python
create_tables(TABLES)
````

````python
import mysql.connector
from mysql.connector import errorcode
````

````python
def insert_from_df(df):    
    """Функция сохранения из csv в MySQL"""
    cnx = mysql.connector.connect(user='USER', database='characters_db')
    cursor = cnx.cursor()

    add_character = ("INSERT INTO characters "
                   "(name, homeworld_name, height, gender, starships_name) "
                   "VALUES (%(name)s, %(homeworld_name)s, %(height)s, %(gender)s, %(starships_name)s)")

    for index, row in df.iterrows():
        data_character = {
                'name': row['name'],
                'homeworld_name': row['homeworld.name'],
                'height': row['height'],
                'gender': row['gender'],
                'starships_name': row['starships.name']
            }
        cursor.execute(add_character, (data_character))

    cnx.commit()
    print("INSERTED")
    cursor.close()
    cnx.close()
````

````python
insert_from_df(characters_df)
````
